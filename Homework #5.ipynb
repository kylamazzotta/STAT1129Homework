# Question #1

Use R to create the following two matrices and do the indicated matrix multiplication. What is the resulting matrix?

```r
matrix1 <- matrix(c(7, 2, 9, 4, 12, 13), nrow = 2, ncol = 3)

matrix2 <- matrix(c(1, 2, 3, 7, 8, 9, 12, 13, 14, 19, 20, 21), nrow = 3, ncol = 4)

matrix1
matrix2
```

What is the resulting matrix?

```r
matrix_multiply <- matrix1 %*% matrix2

matrix_multiply
```

# Question #2d

Plot a bar chart to show the salary of row 1,4,5 and label the bar with people’s name correspondingly. 

```r
x <- c("Peter", "Gary", "Michelle")
y <- c(623.30, 729.00, 843.25)

barplot(y, names.arg = x, col = "orange", density = 50)
```

# Question #2e

Plot a pie chart, based on this data framework, only showing the highest salary, lowest salary, and
median salary, give your plot a title, and label each pie.

```r
max_salary <- max(New_Column_DF$salary)

min_salary <- min(New_Column_DF$salary)

median_salary <- median(New_Column_DF$salary)
```

```r
q = c(max_salary, min_salary, median_salary)

my_label <- (c("Maximum Salary", "Minimum Salary", "Median Salary"))

colors <- c("pink", "purple", "light blue")

pie(q, label = my_label, main = "Salaries", col = colors)

legend("topleft", my_label, fill = colors)
```
