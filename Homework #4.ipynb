# Question #5

Read each student’s grade from the marks dictionary. Then plot a bar chart for each student’s
grade. Label student’s name at each plot. Add plot title. Add legend for the pie chart. 

```python
marks = {'Andy':88,'Amy':66,'James': 90,'Jules': 55,'Arthur': 77}

import matplotlib.pyplot as plt
import numpy as np

names = list(marks.keys())
grades = list(marks.values())

plt.title("Student Names and Grades")
plt.xlabel("Student Name")
plt.ylabel("Student Grade")

plt.bar(names, grades)
plt.show()
```

Read each student’s grade from the marks dictionary. Then plot a pie chart for each student’s grade. Label student’s name at each plot. Add plot title. Add legend for the pie chart.

```python
marks = {'Andy':88,'Amy':66,'James': 90,'Jules': 55,'Arthur': 77}

import matplotlib.pyplot as plt
import numpy as np

grades = list(marks.values())

mylabels = ["Andy", "Amy", "James", "Jules", "Arthur"]

plt.title("Student Names and Grades")

plt.pie(grades, labels = mylabels)
plt.legend(title = "Student Name", bbox_to_anchor = (1.05, 1.0), loc = "upper left")

plt.show()
```

# Question #6

Practice subplots() function, draw a subplot with 2 rows and 3 columns. Add each plot title.
• One subplot has multiple lines,
• one subplot is a bar chat,
• one subplot is a pie chat,
• one subplot has a grid, you define it yourself,
• one subplot is a scatter plot,
• one subplot is a histogram plot. 

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
x1 = np.array([0, 1, 2, 3])
y1 = np.array([5, 4, 3, 2])
plt.subplot(2, 3, 1)
plt.plot(x, y, x1, y1)
plt.title("Multiple Lines")

#plot 2:
day = np.array(["M", "T", "W", "R", "F"])
quantity = np.array([34, 21, 85, 72, 64])
plt.subplot(2, 3, 2)
plt.bar(day, quantity)
plt.title("Favorite Days : A Bar Chart")
plt.xlabel("Day")
plt.ylabel("Quantity")

#plot 3:
frequency = np.array([33, 40, 12, 67, 51])
mylabels = ("Banana", "Strawberry", "Mango", "Peach", "Kiwi")
plt.subplot(2, 3, 3)
plt.pie(frequency, labels = mylabels)
plt.title("Favorite Fruits : A Pie Chart")

#plot 4:
xp = np.array([0, 1, 2, 3, 7])
yp = np.array([33, 8.1, 11, 100, 22])
plt.subplot(2, 3, 4)
plt.plot(xp, yp)
plt.grid()
plt.title("Subplot With A Grid")

#plot 5:
s = np.array([4, 3, 3.5, 15, 7.2, 6.3, 7, 23, 9, 5, 3, 7.7, 3.2, 2.2, 6.4, 8.3, 9.9])
r = np.array([33, 8.1, 11, 100, 22, 64, 3.3, 6.5, 88, 12, 43, 5.5, 23, 95, 34, 5.9, 51])
plt.subplot(2, 3, 5)
plt.scatter(s, r)
plt.title("Scatterplot")

#plot 6:
b = np.random.normal(20, 10, 350 )
plt.subplot(2, 3, 6)
plt.hist(b)
plt.title("Histogram")

plt.show()
```
