# Question #2

```python
import pandas as pd

df = pd.read_csv('amazon-orders.csv')
print(df.head(59))
```

```python
df = df.fillna(0)
print(df.head(59))
```

```python
df["Total Charged"] = df["Total Charged"].str.replace("$","").astype(float)
df.head(59)
```

```python
df["Tax Charged"] = df["Tax Charged"].str.replace("$","").astype(float)
df.head(59)
```

```python
df["Subtotal"] = df["Subtotal"].str.replace("$","").astype(float)
df.head(59)
```

```python
df["Total Charged"].sum()
```

```python
df["Total Charged"].mean()
```

```python
df["Total Charged"].median()
```

```python
df["Total Charged"].max()
```

```python
df["Total Charged"].min()
```

```python
import statistics

statistics.stdev(df["Total Charged"])
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df.head(59)
```

```python
daily_orders = df.groupby("Order Date").sum()["Total Charged"]
daily_orders.head(59)
```

```python
daily_tax = df.groupby("Order Date").sum()["Tax Charged"]
daily_tax.head(59)
```

```python
daily_subtotal = df.groupby("Order Date").sum()["Subtotal"]
daily_subtotal.head(59)
```

```python
daily_orders.plot.bar(figsize = (45, 20), color = "#FC5A50")
font1 = {'family':'serif','color':'green','size':70}
font2 = {'family':'serif','color':'green','size':50}
plt.title("Total Charged on Amazon Each Day", fontdict = font1)
plt.ylabel("Total Charged", fontdict = font2)
plt.xlabel("Order Date", fontdict = font2)
plt.xticks(fontsize=35)
plt.yticks(fontsize=35)

plt.show()
```

```python
df.resample(rule='M', on= "Order Date")["Total Charged"].sum()
```

```python
import matplotlib.pyplot as plt

monthly_orders = df.resample(rule='M', on= "Order Date")["Total Charged"].sum()
plt.pie(monthly_orders, labels = ["April", "May", "June", "July"])
plt.show()
```

```python
import matplotlib.pyplot as plt
import datetime
import numpy as np

#plot 1:
Q = df["Total Charged"].mean()
S = daily_orders

plt.subplot(1, 2, 1)
plt.axhline(Q, color = "green", label = "Average Total Charged")
plt.plot(S, color = "salmon", label = "Daily Order Charged")
plt.xlim([datetime.date(2019, 4, 1), datetime.date(2019, 7, 31)])
plt.xticks(rotation = 45)
plt.legend(loc = 2, fontsize = 7)
plt.title("Daily Charge Compared to Average Total Charge")
plt.xlabel("Date of Order")
plt.ylabel("Total Charged ($)")

#plot 2:
R = df["Tax Charged"].mean()
T = daily_tax

plt.subplot(1, 2, 2)
plt.axhline(R, color = "red", label = "Average Tax Charged")
plt.plot(T, color = "purple", label = "Daily Tax Charged")
plt.xlim([datetime.date(2019, 4, 1), datetime.date(2019, 7, 31)])
plt.legend(loc = 2, fontsize = 7)
plt.title("Daily Tax Compared to Average Tax Charge")
plt.xlabel("Date of Order")
plt.ylabel("Tax Charged ($)")

plt.subplots_adjust(left=.3,
                    bottom=0.1,
                    right=1.6,
                    top=1,
                    wspace=0.4,
                    hspace=0.2)
plt.xticks(rotation = 45)

plt.legend(loc = 2, fontsize = 7)
 
plt.show()
```
